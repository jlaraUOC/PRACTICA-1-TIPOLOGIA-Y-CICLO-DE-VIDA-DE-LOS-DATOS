```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el conjunto de datos
df = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/election.csv")
print(df.head())
# Crear un grÃ¡fico circular con porcentajes
plt.figure(figsize=(8, 8))
counts = df["winner"].value_counts()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title('Porcentaje de Resultados Electorales')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Generar un conjunto de datos aleatorio (en este ejemplo, distribuciones normales)
data = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/diabetes-vid.csv")  # Puedes cargar un dataset de ejemplo o utilizar tu propio dataset

# Mostrar la estructura del dataset (opcional)
print(data.head())

plt.figure(figsize=(8, 6))
sns.violinplot(x="Outcome", y="Glucose", data=data)
plt.title('DistribuciÃ³n de los niveles de glucosa entre grupos')
plt.xlabel('supervivientes o defunciones')
plt.ylabel('glucosa en sangre')
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np

url = 'https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/mtcars.csv'
df = pd.read_csv(url)
df = df.set_index('model')
df = df.reset_index(drop=True)
df.head()

Z = linkage(df, 'ward')
plt.xlabel('sample index')
plt.ylabel('distance (Ward)')

dendrogram(Z, labels=df.index, leaf_rotation=90)

plt.show()
```
